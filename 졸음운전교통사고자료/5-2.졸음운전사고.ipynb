```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams["font.family"]="Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"]= False
```

```python
df1 = pd.read_csv("2014년 졸음운전 교통사고.csv", encoding='euc-kr') #www.data.go.kr
df2 = pd.read_csv("2015년 졸음운전 교통사고.csv", encoding='euc-kr')
df3 = pd.read_csv("2016년 졸음운전 교통사고.csv", encoding='euc-kr')
```

```python
df3
```

```python
mdf = pd.concat( [df1,df2,df3])
mdf
```

```python
# 2. 년도, 월별로 인덱스를 설정(멀티인덱스) 하고
# 데이터를 보여 주시요
```

```python
mdf['년도'] = mdf['구분'].str[0:5]
```

```python
mdf['월'] =mdf['구분'].str[5:]
```

```python
mdf.drop( columns=['구분'], inplace=True)
```

```python
mdf = mdf[['년도','월','사고(건)','사망(명)','부상(명)'] ]
```

```python
mdf.set_index( ['년도','월'] )
```

```python
# 3. 년도 및 월별 평균 사망자를 보여주시요
g= mdf.groupby( ['년도','월'],sort=False)
g.mean()
```

```python
# 4. 2016년 사고대비 사망율을 보여 주시요
mdf[ mdf['년도']=='2016년']
```

```python
##멀티인덱스로 설정한경우
mdf.set_index( ['년도','월'] ,inplace=True)
```

```python
mdf2016 = mdf.loc['2016년']
mdf2016['사망율'] = mdf2016['사망(명)']/mdf2016['사고(건)']
mdf2016
```

```python
# 5. 2014년도 월별 사망, 부상 데이터를 바차트로
# 보여주시요
```
