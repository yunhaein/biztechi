```python
sdf['year']=sdf.index.year
```

```python
sdf['month']=sdf.index.month
```

```python
g = sdf.groupby('year')
g.mean()
```

```python
g = sdf.groupby( sdf.index.year)
g.mean()
```

```python
g = sdf.groupby( sdf.index.month)
g.mean()
```

```python
g = sdf.groupby( [sdf.index.year, sdf.index.month] )
g.mean()
```

```python
# http://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html
```

```python
g = sdf.resample(rule='A')
g.mean()
```

```python
g = sdf.resample(rule='w')
g.mean()
```

```python
g = sdf.resample(rule='Q')
g.mean()
```
